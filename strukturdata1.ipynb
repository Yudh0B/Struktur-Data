persamaan garis lurus -- regresi linier
y = m.x + c

```python
from scipy import stats
#menentukan persamaan y = mx + c
# m : kemiringan
# c : intercept / koefisien
m, c, r, p, err = stats.linregress(x,y)
```

```python
m
```

```python
c
```

```python
def fungsi(x):
    return m*x + c

mdl = list(map(fungsi, x))

plt.scatter(x,y)
plt.plot(x, mdl)
plt.show()
```

```python
#multiple regression : 
#engine size dan konsumsi bbm untuk menentukan harga
z = dm[['engine-size','city-mpg']]
z
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(z,y)

prediksi=reg.predict([[220,20]])
prediksi
```

```python
from sklearn import svm
regr = svm.SVR()
regr.fit(z,y)

prediksi=regr.predict([[220,20]])
prediksi
```
